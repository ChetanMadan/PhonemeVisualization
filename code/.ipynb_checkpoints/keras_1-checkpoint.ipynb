```python
import numpy as np
from random import randint
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt


from sklearn.metrics import confusion_matrix
import itertools

%matplotlib inline
```

```python
train_labels = []
train_samples = []
```

# Data Generation :

## Individuals ranging from 13 - 100 years old.

## 50% young(<65)

### we know from the dataset that 95% of young people dont face side effects and 95% of old people face side-effects

## 0 : No side effect , 1 : Side effect.


```python

# Effected by side effects
for i in range(50):
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(1)
    
    
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(0)


# Not effected by side effects.
for i in range(1000):
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)
    
    
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1)
    
```

```python
#As Keras takes only numpy arrays as inputs converting lists to np arrays

train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
```

```python
#Scaling data

scalar = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scalar.fit_transform(train_samples.reshape(-1,1)) 
```

# Keras 

```python
import keras
from keras import backend as k
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

```python
model = Sequential([
    
    Dense(16,input_shape=(1,), activation='relu'),
    
    Dense(32, activation='relu'),
    
    Dense(2, activation='softmax')
    
    
])
```

```python
model.summary()
```

```python
model.compile(Adam(lr= 0.0001), loss='sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
confusionMatrix = confusion_matrix(test_labels,rounded_predictions)
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title="Confusion Matrix",
                          cmap=plt.cm.Blues):
    fig, ax = plt.subplots()
    im = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    ax.figure.colorbar(im, ax=ax)
    # We want to show all ticks...
    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           # ... and label them with the respective list entries
           xticklabels=classes, yticklabels=classes,
           title=title,
           ylabel='True label',
           xlabel='Predicted label')

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i, j], fmt),
                    ha="center", va="center",
                    color="white" if cm[i, j] > thresh else "black")
    fig.tight_layout()
    return ax
```

```python
cm_plot_labels = ['No Side effects', 'Side effects']
plot_confusion_matrix(confusionMatrix, cm_plot_labels,title='confusion matrix')
```
